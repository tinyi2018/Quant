```python
def Cal_ma5():
    for i in df.index:
        if( i < 4):
            continue
        df2 = df[(i-4):(i+1)]
        df.loc[i,"ma5"] = df2["close"].mean()
    else:
         print(i)
```

```python
def Cal_ma10():
    for i in df.index:
        if( i < 9):
            continue
        df2 = df[(i-9):(i+1)]
        df.loc[i,"ma10"] = df2["close"].mean()
    else:
         print(i)
```

```python
def Cal_ma5_ma10():
    df["ma5"] = 0.0
    df["ma10"] = 0.0
    df["ma5"] = df["close"].rolling(5).mean()
    df["ma10"] = df["close"].rolling(10).mean()
```

```python
def Cal_ma5_ma10_2():
    sr = df["close"].cumsum()
    df["ma5"] = (sr - sr.shift(1).fillna(0).shift(4))/5
    df["ma10"] = (sr - sr.shift(1).fillna(0).shift(9))/10
```

```python
def Cal_ma5Lma10():
    df["ma5>ma10"] = df["ma5"]>df["ma10"]
```

```python
def Cal_ma5Lma10_pre():
    for i in df.index:
        if ( i== 0):
            continue
        df.loc[i,"ma5>ma10_pre"] = (df.loc[(i-1),"ma5"] > df.loc[(i-1),"ma10"]  )
```

```python
def Cal_ma5Lma10_pre_2():
    df["ma5>ma10_pre"] = df["ma5>ma10"].shift(1)
```

```python
 def Cal_jc(): 
    for i in df.index:
         if (i<9):
            print (" i : "+str(i))
            continue
         df.loc[i,"jc"] =(( df.loc[i,"ma5>ma10"] == True ) & (df.loc[i,"ma5>ma10_pre"] == False))
    else:
         print( " Else i = "+ str(i))
```

```python
df["sicha"] = False
def Cal_sc():
    for i in df.index:
         if (i<9):
            print (" i : "+str(i))
            continue
         df.loc[i,"sicha"] =(( df.loc[i,"ma5>ma10"] == False ) & (df.loc[i,"ma5>ma10_pre"] == True))
    else:
         print( " Else i = "+ str(i))
```
